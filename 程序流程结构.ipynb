# **选择结构**

## **if语句**

**作用:** 执行满足条件的语句

**if语句的三种形式:**

- 单行格式if语句: `if(条件){ 条件满足执行的语句 }`
  
- 多行格式if语句: `if(条件){ 条件满足执行的语句 }else{ 条件不满足执行的语句 };`
  
- 多条件的if语句: `if(条件1){ 条件1满足执行的语句 }else if(条件2){条件2满足执行的语句}... else{ 都不满足执行的语句}`

**嵌套if语句:** 在if语句中，可以嵌套使用if语句，达到更精确的条件判断

**参考代码1:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //单行if
    int a;
    cout << "Please input a number: " << endl;
    cin >> a;
    if(a){
        cout << "a > 0" << endl;
    }
    
    return 0;
}
```
**参考代码2:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //多行if
    int a;
    cout << "Please input a number: " << endl;
    cin >> a;
    if(a > 0){
        cout << "a > 0" << endl;
    }
    else{
        cout << "a <= 0" << endl;
    }
    
    return 0;
}
```
**参考代码3:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //多条件if
    int a;
    cout << "Please input a number: " << endl;
    cin >> a;
    if(a > 0){
        cout << "a > 0" << endl;
    }
    else if(a == 0){
        cout << "a = 0" << endl;
    }
    else{
        cout << "a < 0" << endl;
    }
    
    return 0;
}
```
**参考代码4:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //嵌套if
    int a;
    cout << "Please input a number: " << endl;
    cin >> a;
    if(a > 0){
        if(a % 2 == 0){
            cout << "a > 0 and a is even" << endl;
        }
        else{
            cout << "a > 0 and a is odd" << endl;
        }
    }
    else if(a == 0){
        cout << "a = 0" << endl;
    }
    else{
        cout << "a < 0" << endl;
    }
    
    return 0;
}
```

## **三目运算符**

**作用:** 通过三目运算符实现简单的判断

**语法:** `表达式1 ? 表达式2 : 表达式3`

**解释:**

- 如果表达式1的值为真，执行表达式2，并返回表达式2的结果

- 如果表达式1的值为假，执行表达式3，并返回表达式3的结果

**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //三目运算符
    int a, b, c;
    cout << "Please input two numbers: " << endl;
    cin >> a >> b;
    c = a > b ? a : b;
    cout << "Maximum = " << c << endl;
    
    return 0;
}
```

## **switch语句**

**作用:** 执行多条件分支语句

**语法:**
```C++
switch(表达式){

    case 结果1：执行语句;break;

    case 结果2：执行语句;break;

    ...

    default:执行语句;break;

}
```
**注意:**
- switch语句中表达式类型只能是整型或者字符型
- case里如果没有break，那么程序会一直向下执行即执行所有的符合条件的case

**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //swicth
    int a;
    cout << "Please input a number: " << endl;
    cin >> a;
    switch(a){
        case 1 : cout << "a = 1" << endl;break;
        case 2 : cout << "a = 2" << endl;break;
        case 3 : cout << "a = 3" << endl;break;
        default : cout << "a > 3" << endl;break;
    }
    
    return 0;
}
```

# **循环结构**

## **while循环语句**

**作用:** 满足循环条件，执行循环语句

**语法:** ` while(循环条件){ 循环语句 }`

**解释:** 只要循环条件的结果为真，就执行循环语句

**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //while
    int num = 0;
    while(num < 10){
        cout << num++ << endl;
    }
    cout << num << endl;
    
    return 0;
}
```

## **do...while循环语句**

**作用:** 满足循环条件，执行循环语句

**语法:** `do{ 循环语句 } while(循环条件);`

**注意:** 与while的区别在于do...while会先执行一次循环语句，再判断循环条件

**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //do...while
    int num = 1;
    do{
        cout << num++ << endl;
    }while(num == 0);
    cout << num << endl;
    
    return 0;
}
```

## **for循环语句**

**作用:** 满足循环条件，执行循环语句

**语法:** `for(起始表达式;条件表达式;末尾循环体) { 循环语句; }`

**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //for
    for(int i = 0; i < 10; i++){
        cout << i << endl;
    }
    
    return 0;
}
```

**参考代码:**
```C++
#include<iostream>
#include <iomanip>
using namespace std;
int main(void){
    //循环与if综合
    for(int i = 1; i < 10; i++){
        for(int j = 1; j <= i; j++){
            cout << setw(2) << left
                 << i 
                 << setw(2) << left
                 << "*" 
                 << setw(2) << left
                 << j
                 << setw(2) << left
                 << "=" 
                 << setw(4) << left
                 << i * j;
            if(i == j){
                cout << endl;
            }
        }
    }
    
    return 0;
}
```

# **跳转语句**

## **break语句**

**作用:** 用于跳出选择结构或者循环结构

**break的使用场景:**

- 出现在switch条件语句中，作用是终止case并跳出switch
- 出现在循环语句中，作用是跳出当前的循环语句
- 出现在嵌套循环中，跳出最近的内层循环语句

**switch条件语句略**

**参考代码1:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //break在循环中的应用
    int num = 0;
    while(1){
       if(num++ == 10){
           break;
       } 
    }
    cout << "num = " << num << endl;
    
    return 0;
    
}
```
**参考代码2:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //break在嵌套循环中的应用
    for(int i = 0; i < 10; i++){
        for(int j = 0;; j++){
            if(j == i){break;}
            cout << "*";
        }
        cout << endl;
    }
    
    return 0;
    
}
```

## **continue语句**

**作用:** 在循环语句中，跳过本次循环中余下尚未执行的语句，继续执行下一次循环


**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //continue语句
    for(int i = 0; i < 10; i++){
        if(i % 2 == 1){continue;}
        cout << i << endl;
    }
    return 0;
    
}
```

## **goto语句**

**作用:** 无条件跳转语句

**语法:** `goto 标记;`
 
**解释:** 如果标记的名称存在，执行到goto语句时，会跳转到标记的位置

**参考代码:**
```C++
#include<iostream>
using namespace std;
int main(void){
    //go to语句
    int num = 0;
    while(1){
       if(num++ == 10){
           goto flag;
       } 
    }
    flag:
    cout << "num = " << num << endl;
    
    return 0;
    
}
```
